## Task2 -> Data Cleaning

```python
# counting the missing values
print("the number of missing values: ")
df.isnull().sum()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median(), inplace=True)
```

```python
# # Drop cabin column (too many missing values)
# df = df.drop(columns=["Cabin"], inplace= True)
df.head()
```

## Task3: Simple Analysis

```python
avg = df['Survived'].mean()
percentage = avg*100
percentage
```

```python
# Did women survive more than men?
percentage_women = df[df["Sex"]== "female"]["Survived"].mean()*100
percentage_men = df[df["Sex"] == "male"]["Survived"].mean()*100
if(percentage_women > percentage_men):
    print("More number of women survived than men")
elif(percentage_women == percentage_men):
    print("Equal number of men and women had survived")
else:
    print("More number of men survived than women")
```

```python
# Which passenger class had highest survival?
highest_suvival_pclass = df.groupby("Pclass")["Survived"].mean()
highest_suvival_pclass
```

## Task4: Filtering

```python
# Age < 18
temp = df[df["Age"] < 18]
temp.head()
```

```python
# Paid fare > 100
temp2 = df[df.Fare > 100]
temp2.head()
```

```python
# Travelling in 1st class
temp3 = df[df["Pclass"] == 1]
temp3.head()
```

## Task5: Feature Creation

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
percentage_family_survived = df[df["FamilySize"] > 0]["Survived"].mean()*100

percentage_nofam_survived = df[df["FamilySize"]==0]["Survived"].mean()*100

if(percentage_family_survived > percentage_nofam_survived):
    print("more people in families survied")
else:
    print("More people travelling alone had survived")
```

## Task6: Sorting and Insights

```python
# Top 10 highest fares
top_10_highest_fares = df.Fare.sort_values(ascending=False).head(10)
top_10_highest_fares
```

```python
# Oldest passengers
oldest_passengers = df[df.Age > 60]
oldest_passengers.head()
```
